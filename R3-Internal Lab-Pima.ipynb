# Q4 Replace all the 0s in the column with the median of the same column value accordingly.

```python
pima_df_new = pima_df
pima_df_new['Glucose'] = pima_df['Glucose'].replace(0, pima_df['Glucose'].median())
pima_df_new['BloodPressure'] = pima_df['BloodPressure'].replace(0, pima_df['BloodPressure'].median())
pima_df_new['SkinThickness'] = pima_df['SkinThickness'].replace(0, pima_df['SkinThickness'].median())
pima_df_new['Insulin'] = pima_df['Insulin'].replace(0, pima_df['Insulin'].median())
pima_df_new['BMI'] = pima_df['BMI'].replace(0, pima_df['BMI'].median())
```

# Q8 Split the pima_df into training and test set in the ratio of 70:30 (Training:Test).</b>


```python
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation

feature_cols = ['Pregnancies', 'Insulin', 'BMI', 'Age','Glucose','BloodPressure','DiabetesPedigreeFunction','SkinThickness']
X = pima_df_new[feature_cols] # Features
y = pima_df_new.Outcome # Target variable

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
# Create Decision Tree classifer object



```

# Q9 Create the decision tree model using “entropy” method of reducing the entropy and fit it to training data.</b>



```python
clf = DecisionTreeClassifier(criterion='entropy')
# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)
#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

# Q10 Print the accuracy of the model & print the confusion matrix 

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
